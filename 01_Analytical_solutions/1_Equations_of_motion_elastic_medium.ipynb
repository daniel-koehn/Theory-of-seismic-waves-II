# Seismic wave propagation in elastic media

I will shortly review the eqations of motion to describe wave propagation in elastic media. A detailed derivation of the governing equations is described in chapter 2 of the Lecture notes:

["Theory of elastic waves" by Gerhard Müller](https://www.geo.uni-hamburg.de/geophysik/studium/bsc-geophysik-ozeanographie/module/5-sem/vgsw/seismicwaves-mueller2013.pdf)

## Elastic equations of motion

We can describe the **conservation of momentum** in a continuous medium by

\begin{equation}
\rho \frac{\partial^2 u_i}{\partial t^2} = \sum_{j=1}^{3} \frac{\partial}{\partial x_j} \sigma_{ij} + f_i
\end{equation}

with

\begin{align}
u_i &= displacement, \nonumber\\
\sigma_{ij} &= stress\; tensor, \nonumber\\
t &= time, \nonumber\\
x_i &= spatial\; coordinates\; (x_1,x_2,x_3)\; :=\; (x,y,z), \nonumber\\
\rho &= density. \nonumber\\
\end{align}

This system of partial differential equations simply states that we can change the momentum in the medium (LHS) using surface forces described by the stress tensor $\sigma_{ij}$ or body forces $f_i$ (RHS).

These equations are independent of the medium state, i.e. if it is a liquid, solid or gas. Therefore, additional equations are required to describe how the medium is deformed, when a certain amount of stress is applied to the medium.

At a point far away from the source location, we can assume small displacements $u_i$. In this case the relation between the stress $\sigma_{ij}$ and medium deformation $\epsilon_{ij}$ is governed by a linear relationship known as the **generalized Hookes law**:


\begin{equation}
\sigma_{ij} = \sum_{k=1}^{3}\sum_{l=1}^{3} c_{ijkl}\; \epsilon_{kl},
\end{equation}


where the deformation tensor 


\begin{equation}
\epsilon_{ij} = \frac{1}{2}\biggl(\frac{\partial u_i}{\partial x_j} + \frac{\partial u_j}{\partial x_i}\biggr) \nonumber
\end{equation}


describes the deformation state. The 4th order **elastic tensor** $\mathbf c_{ijkl}$ has 81 (= 9 x 9 components). Due to the symmetry of the stress- and strain tensors only 36 (=
6 x 6) components are independent. Since the elastic deformation energy (= elastic
energy per volume unit) is a state function, the number of components can be further reduced to 21 (Aki and Richards (1980), p.21-23). This is the maximum number of independent elasticity constants, which an anisotropic elastic medium can have.

To introduce a simpler notation, I neglected the fact that the wavefields $v_i$, $\sigma_{ij}$ depend on the spatial coordinate $x_i$ and time $t$ in the above equations. The material parameters $\rho$, $c_{ijkl}$ are assumed in most cases only depend on spatial, but no temperal variations. 

This simple system of partial differential equations (1) + (2), together with certain initial and boundary conditions, is all we need to describe the propagation of a seismic wavefield in an elastic medium. This includes, body waves, surface waves, (multiple) reflected waves, refracted waves, diffracted waves, P-S or S-P converted waves ... if I have to bet who can predict the arrival time and amplitude of a seismic wavefield more accurately, a physicist or a fortuneteller, I would bet on the physicist. Think about it for a moment, eq. (1) + (2) allows to predict the future!

The only requirement is that we know the distribution of the density $\rho$ and elastic tensor components $c_{ijkl}$ in the sub-surface. Then we can solve the partial differential equations either analytical or numerically. You will learn how to do this in the upcoming lectures.

### Topics of the lecture

To implement all topics described in [Lecture 0 - Overview](https://danielkoehnsite.files.wordpress.com/2018/04/lect_00_overview_tew2.pdf),
we actually only need to modify some small parts of eqs. (1) + (2):

* To solve eqs. (1) + (2) numerically we have to discretize the material parameters and wavefields on a Cartesian grid and replace the partial differential operators by **Finite Difference** operators 
* **Seismic anisotropy** is already described by the elastic tensor $c_{ijkl}$ , we only have to understand the impact of the different tensor components on the seismic wavefield
* To implement **visco-elastic damping**, we need to modify eq.(2), the elastic tensor components and add some additional partial differential equations
* To model **frequency domain wavefields**, we can either apply a discrete Fourier transform to the modelled time domain wavefields, or we fourier transform eq.(1)
* **Global seismic wave propagation** requires a transformation from the Cartesian to a spherical coordinate system. This affects only the spatial derivatives in eq.(1) and the deformation tensor $\epsilon_{ij}$, but not eq.(2)


## Some other notations

Physicists are known to be lazy, so I introduce some simpler notations for eq.(1) + (2). First, we can get rid of the summations by introducing the [Einstein notation](https://en.wikipedia.org/wiki/Einstein_notation), where repeated indices are assumed as summation. For example, the product of a (m x n) matrix $a_{ij}$ with a (n x 1) column vector $x_j$

\begin{align}
g_i = \sum_{j=1}^n a_{ij} x_j\; (i=1,2,...,m)\nonumber
\end{align}

can be written with Einstein notation as 

\begin{align}
g_i = a_{ij} x_j\nonumber
\end{align}

Therefore, we can write eqs. (1) + (2) as

\begin{align}
\rho \frac{\partial^2 u_i}{\partial t^2} &= \frac{\partial}{\partial x_j} \sigma_{ij} + f_i \nonumber \\
\sigma_{ij} &= c_{ijkl}\; \epsilon_{kl} \nonumber
\end{align}

where the Einstein notation has to be applied twice in the stress-strain relationship. Alternatively, we could use vector notation:

\begin{align}
\rho \frac{\partial^2 \mathbf{u}}{\partial t^2} &= \mathbf{\nabla} \cdot \mathbf{\sigma} + \mathbf{f} \nonumber \\
\mathbf \sigma &= \mathbf{c} : \mathbf \epsilon \nonumber \\
\mathbf \epsilon &= \frac{1}{2}\biggl(\mathbf{\nabla\; u} + (\mathbf{\nabla\; u})^T\biggr) \nonumber
\end{align}
